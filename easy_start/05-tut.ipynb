```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
```

```python
spark = (
    SparkSession
    .builder
    .appName("Test write parquet")
    .getOrCreate()
)
```

```python
parquet_file_path = "data/browser_events.jsonl"
df = spark.read.json(parquet_file_path)
```

```python
write_path = "data/browser_parquet/untitled.txt"
df.write.parquet(write_path)
```
