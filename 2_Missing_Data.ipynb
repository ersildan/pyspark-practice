```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName('missing_data') \
    .getOrCreate()
```

```python
df = spark.read.csv('containsNull.csv', header=True, inferSchema=True)
df.show()
df.printSchema()
```

```python
# df.na.drop(how='any', thresh=None, subset=None)
# how='any' - любой или все ('any', 'all')
# thresh=2 - количество допустимых строк с NUll пропусками
```

```python
df.na.drop().show()
```

```python
df.na.drop(how='all').show()
```

```python
df.na.drop(how='any').show()
```

```python
df.na.drop(subset='Id').show()
```

```python
df.na.drop(subset='Name').show()
```

```python
df.na.drop(thresh=2).show()
```

```python
df.na.drop(how='all').show()
```

```python
df.na.drop(how='all', subset=['Name', 'Sales']).show()
```

```python
# .fill - позволяет заполнить значения
df.na.fill('New values').show()
```

```python
df.na.fill(0).show()
```

```python
# Часто нужно специализировать таблицу или данные
# with + параметр subset
df.na.fill('No Name', subset=['Name']).show()
```

```python
df.na.fill('No data', subset=['Name']).show()
```

```python
from pyspark.sql.functions import mean

mean_val = df.select(
    mean(
        df['Sales']
    )
).collect()

print(mean_val)
```

```python
mean_val[0][0]
```

```python
# Заменяем на медиану по столбцу Sales

mean_sales = mean_val[0][0]
df.na.fill(mean_sales, ['Sales']).show()
```

```python
df.na.fill(df.select(mean(df['Sales'])).collect()[0][0], ['Sales']).show()
```

```python
df.na.fill(
    df.select(
        mean(
            df['Sales']
        )
    )
    .collect()[0][0],
    ['Sales']
).show()
```
